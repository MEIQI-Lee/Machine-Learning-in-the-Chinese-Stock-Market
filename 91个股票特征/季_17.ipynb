```python
import numpy as np
import pandas as pd
```

```python
# 17. chinv
# 季度频率。存货变动除以总资产。

df1 = pd.read_csv('季_资产负债表.csv', usecols=['证券代码', '会计期间', '存货净额', '资产总计'], low_memory=False)
df1['存货比'] = df1['存货净额']/df1['资产总计'].replace(0, np.nan)
df1['chinv'] = df1['存货比'] - df1.groupby('证券代码').shift(1)['存货比']
df2 = df1[['证券代码', '会计期间', 'chinv']]
df2.to_csv('季_17.csv', index=False, encoding='utf-8-sig')
df2
```
