```python
import numpy as np
import pandas as pd
```

```python
# 67.  realestate
# 季度频率。投资性房地产除以固定资产。

df1 = pd.read_csv('季_资产负债表.csv', usecols=['证券代码', '会计期间', '固定资产净额', '投资性房地产净额', '查询成功'], low_memory=False)
# 确保证券代码为整数类型
df1['证券代码'] = pd.to_numeric(df1['证券代码'], errors='coerce').fillna(0).astype(int)

df1['投资性房地产净额'][df1['查询成功'].astype('bool')] = df1['投资性房地产净额'][df1['查询成功'].astype('bool')].fillna(0)
df1['realestate'] = df1['投资性房地产净额']/df1['固定资产净额'].replace(0, np.nan)
df2 = df1[['证券代码', '会计期间', 'realestate']]
df2.to_csv('季_67.csv', index=False, encoding='utf-8-sig')
df2
```
