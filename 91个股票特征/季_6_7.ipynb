```python
import pandas as pd
```

```python
# 06. bm
# 季度频率。账面市值比，等于权益的账面价值除以市值。

# 07. bm_ia
# 季度频率。经行业调整的账面市值比。

df1 = pd.read_csv('季_相对价值指标.csv', usecols=['证券代码', '会计期间', '账面市值比A'])

# 确保证券代码为整数类型
df1['证券代码'] = pd.to_numeric(df1['证券代码'], errors='coerce').fillna(0).astype(int)

df1.rename({'账面市值比A':'bm'}, axis='columns', inplace=True)

df2 = pd.read_csv('公司文件.csv')[['证券代码', '行业代码C']]
# 确保证券代码为整数类型
df2['证券代码'] = pd.to_numeric(df2['证券代码'], errors='coerce').fillna(0).astype(int)

df3 = pd.merge(df1, df2, on=['证券代码'], how='left')
df4 = df3[['行业代码C', '会计期间', 'bm']].groupby(['行业代码C', '会计期间']).mean().reset_index()
df4.rename(columns={'bm':'bm_i'}, inplace=True)
df5 = pd.merge(df3, df4, on=['行业代码C', '会计期间'], how='left')
df5['bm_ia'] = df5['bm'] - df5['bm_i']
df6 = df5[['证券代码', '会计期间', 'bm', 'bm_ia']]
df6.to_csv('季_6_7.csv', index=False, encoding='utf-8-sig')
df6
```
