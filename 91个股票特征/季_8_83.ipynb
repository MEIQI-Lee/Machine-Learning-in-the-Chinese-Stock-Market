```python
import pandas as pd
```

```python
# 08. cash
# 季度频率。现金及现金等价物除以平均总资产。

# 83. tang
# 季度频率。现金持有 + 0.715 * 应收项目 + 0.547 * 存货 + 0.535 * 固定资产

df1 = pd.read_csv('季_资产负债表.csv', usecols=['证券代码', '会计期间', '现金及存放中央银行款项', '应收账款净额', '存货净额', '固定资产净额', '资产总计', '查询成功'])
df1['证券代码'] = pd.to_numeric(df1['证券代码'], errors='coerce').fillna(0).astype(int)
df1['现金及存放中央银行款项'][df1['查询成功'].astype('bool')] = df1['现金及存放中央银行款项'][df1['查询成功'].astype('bool')].fillna(0)
df1['tang'] = df1['现金及存放中央银行款项'] + 0.715*df1['应收账款净额'] + 0.547*df1['存货净额'] + 0.535*df1['固定资产净额']
df1['cash'] = df1['现金及存放中央银行款项']/df1['资产总计']
df2 = df1[['证券代码', '会计期间', 'cash', 'tang']]
df2.to_csv('季_8_83.csv', index=False, encoding='utf-8-sig')
df2
```
