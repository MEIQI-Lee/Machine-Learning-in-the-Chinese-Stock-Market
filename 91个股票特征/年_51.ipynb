```python
import numpy as np
import pandas as pd
```

```python
# 51. pchcapx_ia
# 年频率。行业调整后的资本支出变动百分比。

df1 = pd.read_csv('季_现金流量表(直接法).csv', usecols=['证券代码', '会计期间', '购建固定资产、无形资产和其他长期资产支付的现金'])
df1['证券代码'] = pd.to_numeric(df1['证券代码'], errors='coerce').fillna(0).astype(int)
df1['年'] = df1['会计期间'].apply(lambda x: True if x[5:7] == '12' else False)
df1 = df1[df1['年']].reset_index().drop(columns=['index'])
df1['pchcapx'] = (df1['购建固定资产、无形资产和其他长期资产支付的现金'] - df1.groupby('证券代码').shift(1)['购建固定资产、无形资产和其他长期资产支付的现金'])/df1.groupby('证券代码').shift(1)['购建固定资产、无形资产和其他长期资产支付的现金'].replace(0, np.nan)
df2 = pd.read_csv('公司文件.csv', usecols=['证券代码', '行业代码C'])
df2['证券代码'] = pd.to_numeric(df2['证券代码'], errors='coerce').fillna(0).astype(int)
df3 = pd.merge(df1, df2, on=['证券代码'], how='left')
df4 = df3[['行业代码C', '会计期间', 'pchcapx']].groupby(['行业代码C', '会计期间']).mean().reset_index()
df4.rename(columns={'pchcapx':'pchcapx_i'}, inplace=True)
df5 = pd.merge(df3, df4, on=['行业代码C', '会计期间'], how='left')
df5['pchcapx_ia'] = df5['pchcapx'] - df5['pchcapx_i']
df6 = df5[['证券代码', '会计期间', 'pchcapx_ia']]
df6.to_csv('年_51.csv', index=False, encoding='utf-8-sig')
df6
```
