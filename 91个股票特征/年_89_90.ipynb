```python
#十大流通股东股权集中文件
#本表数据总记录数：237240 条，数据频率：季度，数据开始时间： 2003-12-31，数据结束时间： 2024-09-30

#largestholderrate 最大股东持有的普通股的百分比。
#top10holderrate 前10大股东持有的普通股比例。

#Negshrcr1 [股权集中指标1(%)] - 公司第一大流通股股东持股比例
#Negshrcr4 [股权集中指标4(%)] - 公司前10位流通股股东持股比例之和
```

```python
import numpy as np
import pandas as pd 

# 读取Excel文件
HLD = pd.read_excel('HLD_NegCR.xlsx')
HLD = HLD.iloc[2:]

# 将持股比例转换为小数
HLD['Negshrcr1'] = HLD['Negshrcr1'] / 100
HLD['Negshrcr4'] = HLD['Negshrcr4'] / 100

# 重命名列
HLD = HLD.rename(columns={
    'Negshrcr1': 'largestholderrate',
    'Negshrcr4': 'top10holderrate',
    'Reptdt':'会计期间',
    'Stkcd':'证券代码'
})

# 日期转换和提取年份
#HLD['Reptdt'] = pd.to_datetime(HLD['Reptdt'], format='%Y-%m-%d')
#HLD['year'] = HLD['Reptdt'].dt.year

# 创建新的DataFrame并去重
HLD = HLD[['证券代码', '会计期间', 'largestholderrate', 'top10holderrate']].reset_index(drop=True)
HLD = HLD.drop_duplicates(subset=['证券代码', '会计期间'], keep='last').reset_index(drop=True)

# 将证券代码转换为整数类型
HLD['证券代码'] = pd.to_numeric(HLD['证券代码'], errors='coerce').fillna(0).astype(int)

# 打印前几行以确认
print(HLD.head())

# 创建会计期间的DataFrame
year_df = pd.DataFrame([f'{y}-12-31' for y in range(2000, 2024)], columns=['会计期间'])
year_df['会计期间'] = year_df['会计期间'].astype(str)  # 确保为字符串类型

# 读取公司文件
stk_df = pd.read_csv('公司文件.csv')

# 添加辅助列以便合并
stk_df['on'] = 0
year_df['on'] = 0

# 合并DataFrames
df = pd.merge(stk_df, year_df, on=['on'], how='outer')[['证券代码', '会计期间']]
df1 = pd.merge(df, HLD, on=['证券代码', '会计期间'], how='left')

# 打印合并后的DataFrame以确认
df1

# 保存最终结果到CSV文件
df1.to_csv("年_89_90.csv", encoding='utf_8_sig', index=False)
```
