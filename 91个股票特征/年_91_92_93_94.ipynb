```python
#下载表名:中国上市公司股权性质文件 ,数据区间 2003-12-31 至 2023-12-31
#年度变量
import numpy as np 
import pandas as pd 

EN = pd.read_excel('EN_EquityNatureAll.xlsx')
EN = EN.iloc[2:]

EN['soe'] = 0
EN['private'] = 0
EN['foreign'] = 0
EN['others'] = 0

EN['EquityNatureID'] = EN['EquityNatureID'].fillna(0)
EN.groupby('EquityNatureID')['EquityNatureID'].count()


def My_contain(Set,Col,Key,FillCol,Num):
    Set.loc[Set[Col].str.contains(Key, na=False), FillCol] = Num

Set = EN
Col = 'EquityNatureID'
Num = 1
My_contain(Set,Col,'1','soe',Num)
My_contain(Set,Col,'2','private',Num)
My_contain(Set,Col,'3','foreign',Num)
My_contain(Set,Col,'4','others',Num)
My_contain(Set,Col,'0','others',Num)

EN_ = EN[['Symbol','EndDate','soe', 'private', 'foreign', 'others']]
EN_ = EN_.rename(columns = {'Symbol': '证券代码','EndDate':'会计期间'})

EN_ = EN_[['证券代码','会计期间','soe', 'private', 'foreign', 'others']].reset_index(drop=True)

EN_['证券代码'] = EN_['证券代码'].astype(int)

# 创建会计期间的DataFrame
year_df = pd.DataFrame([f'{y}-12-31' for y in range(2000, 2024)], columns=['会计期间'])
year_df['会计期间'] = year_df['会计期间'].astype(str)  # 确保为字符串类型

# 读取公司文件
stk_df = pd.read_csv('公司文件.csv')

# 添加辅助列以便合并
stk_df['on'] = 0
year_df['on'] = 0

# 合并DataFrames
df = pd.merge(stk_df, year_df, on=['on'], how='outer')[['证券代码', '会计期间']]
df1 = pd.merge(df, EN_, on=['证券代码', '会计期间'], how='left')

# 打印合并后的DataFrame以确认
df1

#EN_.to_csv("EN_treated.csv",encoding='utf_8_sig',index = False)
```
